### 데이터분석 라이브러리

1. Python의 주요 데이터분석 라이브러리

   1) numpy : 파이썬 데이터분석의 기본적인 기능들을 제공. 특히 벡터 및 행렬연산과 관련된 편리한 기능들 제공해 주는 라이브러리.

   3) matplotlib : 데이터분석결과에 대한 시각화를 빠르고 직관적으로 수행한다.

```python
%matplotlib 
import numpy as np
import pandas as pd
import matplotlib
%who
```

```python
# 50개의 난수를 발생
data = np.random.rand(50)
print(data)
print(type(data))
```

```python
# 1차원 데이터구조 series
# pandas.Series(데이터) : 인덱스와 값을 리턴
s = pd.Series(data)
s
```

```python
# pandas그래프 그리기
s.plot()
```

```python
# 넘파이로 랜덤값을 생성, 10행 3열의 2차원배열(정확하게 행렬)
data_set = np.random.rand(10, 3)
data_set
```

```python
# 2차원 자료구조 : DataFrame
df = pd.DataFrame(data_set)
print(df)

df1 = pd.DataFrame(data_set, columns=['2019', '2018', '2017'])
df1
```

```python
# 2차원 자료로 Line Graph 
# plot()은 기본이 line
#df1.plot?
df1.plot()
```

```python
# bar그래프
df1.plot(kind='bar')
```

```python
# barh그래프
df1.plot(kind='barh')
```

```python
# Area 그래프
df1.plot(kind='area')
```

```python
# Area 그래프 - stacked
df1.plot(kind='area', stacked=False)
```

```python
# pie chart
# 5개의 난수 발생, 시리즈 자료형의 인덱값을 지정
ser = pd.Series(np.random.rand(5), 
                index=['2019','2018','2017','2016','2015'],
                name='Sales Amount')
print(ser)
```

```python
# pi chart
ser.plot(kind='pie', fontsize=20, figsize=(10, 5), autopct='%.2f %%')
```
