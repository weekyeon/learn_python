# Numpy 함수

* random
    - 0~1 사이의 실수 1개를 난수로 리턴
* sum
* mean
    - 통계 관련 함수
* sort
* unique
    - 중복 제거 함수
* zeros(n)
    - n개의 배열 크기만큼 0으로 채워주는 함수
* ones
    - n개의 배열 크기만큼 1로 채움
* arange(n)
    - n개의 배열 크기만큼 1씩 증가

```python
import numpy as np

arr1 = np.random.randn(5, 3)
arr2 = np.random.randn(5, 3)

print(arr1)
print()
print(arr2)
```

### 1개의 Array에 적용되는 함수

```python
 # print(dir(np))

print("-------- zeros --------")
print(np.zeros(10))
print(np.zeros((10, 5)))
print("-----------------------")
print()

print("-------- ones --------")
print(np.ones(10))
print(np.ones((10, 5)))
print("-----------------------")
print()

# random

# np.random?
# ==================== =========================================================
# Compatibility functions
# ==============================================================================
# rand                 Uniformly distributed values.
# randn                Normally distributed values.
# ranf                 Uniformly distributed floating point numbers.
# randint              Uniformly distributed integers in a given range.
# ==================== =========================================================

arr1 = np.random.randn(5, 3)
print(arr1)
print(abs(arr1))
print("-"*50)

# square : 각 요소의 제곱근 계산
print(np.square(arr1))
print("-"*50)

# exp
# 각 요소의 무리수를 e지수로 계산하기
print(np.exp(arr1))
print("-"*50)

# log() : 각 요소를 자연로그
print(np.log(arr1))
print("-"*50)

# log10() : 상용로그
print(np.log10(arr1))
print("-"*50)

# log2() : 밑이 2인 로그
print(np.log2(arr1))
print("-"*50)

# sign() : 각 요소의 값을
# +인 경우 1
# -인 경우 -1
# 0인 경우 0
print(np.sign(arr1))
print("-"*50)
# 얘가 왜 필요하냐면~
# sigmoid? 라는 게 있다 어쩌고저쩌고이러쿵저러쿵쉬는시간주세요
# input이 들어왔을 때 output이 0아니면 1 나오게 처리해야하는 경우가 있음

# ceil() : 정수 중 가장 큰 값 가져옴
np.ceil(arr1)
print("-"*50)

# floor()
np.floor(arr1)
print("-"*50)

# isnan()
np.isnan(arr1)
print("-"*50)

# isinf()
np.isinf(arr1)
print("-"*50)

# cos()
np.cos(arr1)
print("-"*50)

# sin()
np.sin(arr1)
print("-"*50)

# tan()
np.tan(arr1)
print("-"*50)

# cosh()
np.cosh(arr1)
print("-"*50)

# sinh()
np.sinh(arr1)
print("-"*50)

# tanh()
np.tanh(arr1)
```

### 2개의 Array에 적용되는 함수

```python
# 2개의 array 연산함수(사칙연산)
print(np.multiply(arr1, arr2))
print()
print(np.add(arr1, arr2))
```

```python
# maximum, minimum

print(np.maximum(arr1, arr2))
print()
print(np.minimum(arr1, arr2))
```

# 자주 사용하는 통계 함수

* 통계 함수를 통해서 Numpy 배열의 총합 또는 평균들을 구할 때 사용
* 열의 합 또는 행의 합 등을 구할 수 있음

```python
# 합계 : sum()
print(np.sum(arr1)) # arr1 전체
print("-"*50)

print(np.sum(arr1, axis=1)) # arr1 행의 합
print("-"*50)

print(np.sum(arr1, axis=0)) # arr1 열의 합
```

```python
# 평균 : mean()
print(np.mean(arr1)) # arr1 전체
print("-"*50)

print(np.mean(arr1, axis=1)) # arr1 행의 평균
print("-"*50)

print(np.mean(arr1, axis=0)) # arr1 열의 평균
```

```python
# 통계함수 : 표준편차(std), 분산(var), 최솟값(min), 최댓값(max)
print("------------------표준편차------------------")
print(np.std(arr1))
print("-"*50)
print(np.std(arr1, axis=1))
print("-"*50)
print(np.std(arr1, axis=0))
print()
print("------------------분산------------------------")
print(np.var(arr1))
print("-"*50)
print(np.var(arr1, axis=1))
print("-"*50)
print(np.var(arr1, axis=0))
print()
print("------------------최솟값------------------------")
print(np.min(arr1))
print("-"*50)
print(np.min(arr1, axis=1))
print("-"*50)
print(np.min(arr1, axis=0))
print()
print("------------------최댓값------------------------")
print(np.max(arr1))
print("-"*50)
print(np.max(arr1, axis=1))
print("-"*50)
print(np.max(arr1, axis=0))
print()
```

```python
# 위치함수 : 최솟값(argmin), 최댓값(argmax)이 위치한 idx 리턴
print(np.argmin(arr1))
print(np.argmax(arr1))
```

```python
# 누적함수 : 각 성분의 누적합(cumsum), 누적곱(cumprod)
print(np.cumsum(arr1))
print(np.cumprod(arr1))
```

# 기타 함수

```python
# sort() : 배열 정렬

print(arr1)
print("-"*50)
print("-"*50)
print("-"*50)

print(np.sort(arr1)) # 오름차순
print("-"*50)
print(np.sort(arr1)[::-1]) # 내림차순
print("-"*50)
print(np.sort(arr1, axis=1)) # 
```

```python
# unique 함수 : 중복 제거 함수

names = np.array(["사자", "기린", "오리", "토끼", "호랑이", "오리", "토끼", "고양이"])
print(names)
print(np.unique(names))
```
