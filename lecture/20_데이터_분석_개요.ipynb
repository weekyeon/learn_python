# 데이터 분석 라이브러리

* Python의 주요 데이터 분석 라이브러리
    - numpy
        + 파이썬 데이터 분석의 기본적인 기능들 제공

        + Series, DataFrame 등의 자료구조를 활용해서 데이터 분석에 우수한 성능 발휘
        + 대용량의 데이터를 보다 빠른 속도로 처리할 수 있게 지원
    - matplotlib
        + 데이터 분석 결과에 대한 시각화를 빠르고 직관적으로 수행

```python
# !pip list

# 매직명령어 중 시각화관련 명령어
# %matplotlib : 그래프를 즉시, 웹에 출력할 것인지 여부를 결정
# %매직명령어?

%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib
%who
```

```python
# 50개의 난수를 발생
data = np.random.rand(50)
print(data)

# 만약 data를 웹상에서 오픈된 데이터를 가져왔다고 쳐봐!
# 그럼 랜덤을 일단 쓰진 않겠지..
# 1차원 데이터구조 series로 데이터를 만들어요
# 무슨 말인데 이게
```

```python
# 1차원 데이터구조 series
# pandas.Series(데이터) : 인덱스와 값을 리턴

s = pd.Series(data)
s
```

```python
# pandas 그래프 그리기
s.plot()
```

```python
# 넘파이로 랜덤값 생성
# 10행 3열의 2차원 배열(정확하게는 행렬)
data_set = np.random.rand(10, 3)
data_set
```

```python
# 2차원 자료구조 : Data Frame
df = pd.DataFrame(data_set)
df

# 되게 신기하넹
# DataFrame(data=None, index=None, columns=None, dtype=None, copy=False)
# index = 0~9
# columns = 0~1
```

```python
df1 = pd.DataFrame(data_set, columns=['2019', '2018', '2017'])
df1
```

```python
# 2차원 자료로 Line Graph
# df1.plot(*args, **kwargs)
# plot()은 기본이 Line

df1.plot()
```

```python
# bar 그래프
df1.plot(kind="bar")
```

```python
# Area 그래프 stacked
df1.plot(kind="area")
```

```python
# Area 그래프 stacked (누적?)
df1.plot(kind="area", stacked=False)
```

```python
# pie chart
# 5개의 난수 발생, 시리즈 자료형의 인덱스값을 지정
ser = pd.Series(np.random.rand(5),
                index=['2019', '2018', '2017', '2016', '2015'],
                name="판매실적")
ser
```

```python
# pi chart
ser.plot(kind="pie", fontsize=20, figsize=(10, 5), autopct='%.2f%%')
```
