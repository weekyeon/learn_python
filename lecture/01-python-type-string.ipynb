## 문자열 관련 함수

* count()  : 문자열의 문자갯수를 리턴하는 함수 (대소문자 구분함)
* find()   : 특정문자의 위치를 리턴하는 함수
    + 찾는 문자가 없을 경우 -1 리턴
    + 찾는 문자열이 중복될 경우, 처음 문자의 위치를 리턴
* index()  : 특정문자의 위치를 리턴하는 함수
    + 찾는 문자가 없을 경우 <span style="color:red">**Error**</span>
    + 찾는 문자열이 중복될 경우, 처음 문자의 위치를 리턴
* join()   : 문자열 삽입
* upper()  : 소문자 -> 대문자
* lower()  : 대문자 -> 소문자
* lstrip() : 왼쪽 공백 지우기
* rstrip() : 오른쪽 공백 지우기
* strip()  : 양쪽 공백 지우기 (가운데 공백 못 지움)
* replace(): 문자열 바꾸기
* split()  : 문자열 쪼개기

```python
a = "   Life is too short   "
b = "!   Life is too short   !"
print("a.count('o') == ", a.count("o"))
print("="*50)
print("a.find('o') == ", a.find("o"))
print("="*50)
print("a.find('a') == ", a.find("a"))
print("="*50)
print("a.index('o') == ", a.index("o"))
print("="*50)
# print("a.index('a') == ", a.index("a")) # Error
print("a.join('문자열 삽입') == ", a.join("문자열 삽입"))
print("','.join('abcd') == ", ','.join('abcd'))
print("="*50)
print("a.upper() == ", a.upper())
# print("a.upper('t') == ", a.upper('t')) # Error
print("a.lower() == ", a.lower())
print("="*50)
print("a.lstrip() == ", a.lstrip())
print("a.rstrip() == ", a.rstrip())
print("a.strip() == ", a.strip())
print("==== 문자열의 처음이 공백이 아니라면 실행 X ====")
print("b.lstrip() == ", b.lstrip())
print("b.rstrip() == ", b.rstrip())
print("b.strip() == ", b.strip())
print("="*50)
print("a.replace('Life', 'Your leg') == ", a.replace('Life', 'Your leg'))
print("="*50)
print("a.split(' ') == ", a.split(' '))
print("'a,b,c,d'.split(',') == ", 'a,b,c,d'.split(','))
```
